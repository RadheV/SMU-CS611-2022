To process the data, we need the libraries of pandas

```python
import pandas as pd
import pyarrow.parquet as pq
```

Load the csv file 

```python
df = pd.read_csv("gs://zhenjie_image_bucket/books.csv")
df.head(5)
```

Count the number of rows after droping rows with missing values on publisher

```python
df.shape[0]
```

Find missing values in the column date of publication 

```python
df_with_date = df.dropna(subset=['Date of Publication'])
df_with_date.head(5)
#df_with_data.shape[0]
```

Use regular expression to find wrong formats in the colum "date of publication"

```python
regex = "([1][0-9]{3})"

df_with_date['New Date'] = df_with_date['Date of Publication'].str.extract(regex)
```
